## Improper Integrals

Today, we're learning about, practicing with, and exploring __Improper Integrals.__

Below is a Python cell with which we'll explore __Convergence__ & __Divergence.__

```python
%matplotlib inline

import numpy as np
import sympy as sp
from matplotlib.pyplot import plot as plt

# f(x) = 1/x --> Diverges
f = lambda x: 1/x

# g(x) = x^-2 --> Converges
g = lambda x: 1/(x**2)

sp.mpmath.plot([f, g], xlim=[0,10], ylim=[0,3], points=1500)
```

```python
f = lambda x: sp.functions.exp(x) * sp.functions.log(x)

# Axes config
xlim = [-10,10]
ylim = [-10,10]
points = 500

sp.mpmath.plot(lambda x: sp.functions.exp(x), xlim, ylim, points)
```

```python
sp.mpmath.plot(lambda x: sp.functions.elementary.trigonometric.cot(x), xlim=[-10, 10], ylim=[-3, 10], singularities=[-sp.pi, 0, sp.pi])
```
