# Partial Fraction Decomposition 

<p class='lead'>
Partial Fraction Decomposition is an algebraic technique used to facilitate integration of Rational Functions.</p>

<p>It is to Rational Functions what factoring is to Polynomial expressions&mdash;a process for, essentially, "reverse-engineering" a (relatively) complex expression.
</p>

<p>
Using Partial Fraction Decomposition, you're taking apart a rational function (i.e. a fraction), and breaking it down into its constituent parts.
</p>

## Why Do This?
The smaller, more simple fractions that result from partial fraction technique tend to be _much easier_ to integrate than the original rational function you begin with, in most problems.

## Combining Rational Functions
To understand how Partial Fraction Decomposition works, it's helpful to see an example of the operation in reverse: adding to simple fractions together to get a more complex rational function:

$$\text{ Add the fractions: } \\ \frac{1}{x} \text{ and } \frac{1}{x^2+4}$$

$$\frac{1}{x} + \frac{1}{x^2+4} = \frac{1}{x} \cdot \frac{(x^2+4)}{(x^2+4)} + \frac{(x)}{(x)} \cdot \frac{1}{x^2+4}$$

$$= \frac{x^2+x+4}{x(x^2+4)}$$

## Method

<p class='error'><strong>Note:</strong> this method is incomplete. Need to add steps for decomposing rational functions which include higher-order polynomials.<br/>(e.g. $(x-3)^2, (x-6)^3, (x-2)^4, (x^2+4x+1), (x^2+5x-7)^2$, etc.)</p>

To decompose a rational function, you find the original factors which, added together, made the rational function you need to integrate.

__Note:__ In order to use the _Partial Fraction_ technique to write a rational function $\frac{f(x)}{g(x)}$ as the sum of two partial fractions, you must have two things:
1. The __degree of $f(x)$ must be less than the degree of $g(x)$__ (i.e. the fraction must be _proper_).
2. The __factors of $g(x)$ must be known.__
    In theory, any polynomial can be factored. But in practice, it can be very difficult to factor some polynomials.
    
If the degree of the numerator is greater than the denominator, then the fraction isn't _proper_ (the denominator can still be divided into the numerator), so that's what you must do before you can proceed: perform _polynomial long division_ (or synthetic division, if you're into that sort of thing) to reduce the $\text{ degree of } f(x) < g(x)$ before continuing.

---

### Example
The rational function here can be decomposed and re-written as the sum of two more simple fractions:

$$\frac{5x-3}{x^2-2x-3} = \frac{2}{x+1} + \frac{3}{x-3}$$

<small><em>You can verify this equation algebraically by using the common denominator $(x+1)(x-3)$ to re-join them, and then simplifying the numerator.</em></small>

To integration the partial fractions, just integrate each fraction individually and the add the integrals together.

$$\int \frac{5x-3}{x^2-2x-3}dx = \int \frac{2}{x+1}dx + \int \frac{3}{x-3}dx$$

### Steps

To use the method to integrate $\int \frac{5x-3}{x^2-2x-3}dx, follow these steps:

<tr>
    <td style='width: 40%;'>
        __Step 1__
        Break the rational function into its constituent parts by splitting the denominator into its factors. 
    </td>
    <td style='width: 60%;'>
        $$\frac{5x-3}{x^2-2x-3} = \frac{5x-3}{(x+1)(x-3)}$$
    </td>
</tr>
<tr>
    <td style='width: 40%;'>
        __Step 2__
        Use $A$ and $B$ (and successive capital constants as needed) for the numerators, for now.
    </td>
    <td style='width: 60%;'>
        $$\frac{5x-3}{(x+1)(x-3)} = \frac{A}{x+1} + \frac{B}{x-3}$$
    </td>
</tr>
<tr>
    <td colspan='2'>
        __Step 3__
        Before you can find $A$ and $B$, clear the equation of fractions by multiplying both sides by the common denominator, $(x+1)(x-3)$ <br/><br/>
        
        $$(x+1)(x-3) \left( \frac{5x-3}{(x+1)(x-3)} \right) = \left[ \frac{A}{x+1} + \frac{B}{x-3} \right] \cdot (x+1)(x-3)$$
        <br/>
        $$5x-3 = A(x-3) + B(x+1)$$
    </td>
</tr>

<tr>
    <td colspan='2'>
        __Step 4__
        To find $A$ and $B$, substitute an appropriate value into (first) the $(x+1)$ term to remove it from the equation (i.e. to make it equal to $0$), then do the same for $(x-3)$.<br/><br/>
        
        <table class='no-border' style='width: 100%'>
        <tr>
            <td>
                Subtitute $x = -1$:
                $$5(-1)-3 = A[(-1)-3] + B[(-1)+1]$$<br/>
        
                $$-5-3 = A(-4) + B(0)$$<br/>

                $$-4A = -8$$<br/>

                $$A = \frac{-8}{-4}$$<br/>

                $$A = 2$$
            </td>

                $$5(3)-3 = A[(3)-3] + B[(3)+1]$$<br/>
                
                $$15-3 = A(0) + B(4)$$<br/>
                
                $$4B = 12$$<br/>
                
                $$B = \frac{12}{4}$$<br/>
                
                $$B = 3$$
            </td>
